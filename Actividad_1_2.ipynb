Carga de archivo XLSX

```python
#Carga desde un archivo .xslx sin indice
df = pd.read_excel('Datos de FacturacioÃÅn.xlsx')
df.head()
```

```python
df.info()
```

```python
#Reemplazo de datos tipo string a datos tipos num√©ricos
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0'},regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'6781|':'6781'},regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'3612C':'3612'},regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'      ':'0'},regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2385A':'2385'},regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2510A':'2510'},regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'8440Q':'8440'},regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'},regex=False)
```

```python
#Convertimos datos de la columna de tipo Object a Float
df['CVE_CLPV'] = df['CVE_CLPV'].astype(float)
```

```python
#Filtro1
filtro1 = df[(df['CVE_CLPV']>=1000) & (df['CVE_CLPV']<2000)]
filtro1.head()
#Convertir archivo filtrado a CSV
filtro1.to_csv("filtro2.1.csv")
#Descargar archivo filtrado en csv
from google.colab import files
files.download("filtro2.1.csv")
```

```python
#Filtro2
filtro2 = df[~(df['CVE_VEND']==5) & ~(df['CVE_VEND']==4)]
#Verificar
verificar = filtro2.groupby(['CVE_VEND'])['CVE_VEND'].count()
verificar
filtro2.head()
#Convertir archivo filtrado a CSV
filtro2.to_csv("filtro2.2.csv")
#Descargar archivo filtrado en csv
from google.colab import files
files.download("filtro2.2.csv")
```
